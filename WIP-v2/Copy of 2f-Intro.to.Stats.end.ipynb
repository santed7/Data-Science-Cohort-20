## Example from stats lecture

```python
samples = [85, 90, 92, 73, 95, 60, 89, 78, 99, 90]
samples
```

Sum ( add up )

```python
summation = 0
for value in samples:
  summation += value

summation
```

```python
sum(samples)
```

Count ( len of data set )

```python
len(samples)
```

```python
count = len
count(samples)
```

Mean ( average )

```python
summation/len(samples)
```

```python
summation/count(samples)
```

Median ( mid-point )

```python
sorted(samples)
```

```python
sum(sorted(samples)[4:6])/2
```
